```python
import pandas as pd

data = {
    'Store': ['A', 'A', 'A', 'B', 'B', 'C', 'C', 'C'],
    'Month': ['Jan', 'Feb', 'Mar', 'Jan', 'Feb', 'Jan', 'Feb', 'Mar'],
    'Sales': [100, 150, 200, 120, 130, 90, 110, 160]
}

df = pd.DataFrame(data)
print(df)
```

```python
# Sort data by Store and Month to ensure the last entry per Store is meaningful
df_sorted = df.sort_values(by=['Store', 'Month'])



df_sorted
```

```python
# Group by 'Store' and get the last row for each group using slicing
last_rows = df.groupby('Store').apply(lambda x: x.iloc[-1]).reset_index(drop=True)

print(last_rows)
```
