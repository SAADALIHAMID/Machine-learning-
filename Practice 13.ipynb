```python
import pandas as pd
import numpy as np

# Define the data dictionary
data = {
    'ColA': ['2', '3', '4', np.nan],   # '2', '3', '4' are strings
    'ColB': ['3', '6', '7', '8'],      # All values are strings
    'Col3': [1, 2, 3, None]            # Mixed types with None (Pythonic NaN)
}

# Create DataFrame
df = pd.DataFrame(data)

# Convert columns to numeric (handling errors due to NaNs)
df['ColA'] = pd.to_numeric(df['ColA'], errors='coerce')  # Convert to numeric (int/float), NaN remains
df['ColB'] = pd.to_numeric(df['ColB'])  # No NaNs here, conversion succeeds
df['Col3'] = pd.to_numeric(df['Col3'])  # Convert column to numeric

# Calculate mean values, ignoring NaNs
CA = df['ColA'].mean()
C3 = df['Col3'].mean()

# Fill missing values
df['ColA'].fillna(CA, inplace=True)
df['Col3'].fillna(C3, inplace=True)

# Convert to int if no missing values remain
df = df.astype({'ColA': 'int64', 'ColB': 'int64', 'Col3': 'int64'})

print(df)

```

```python
import pandas as pd
import numpy as np

# Define the data dictionary
data = {
    'ColA': ['2', '4', '4', '2','2'],   # '2', '3', '4' are strings
    'ColB': [np.nan, '6', '6', np.nan,'7'],      # All values are strings
    'Col3': [1, 2, 3, None,5]            # Mixed types with None (Pythonic NaN)
}

df=pd.DataFrame(data)
print(df)

knni=KNNImputer(n_neighbors=2)
new_Data=knni.fit_transform(df)
print('KNN Imputation')
print(new_Data)

lr=IterativeImputer(estimator=LinearRegression())
new_Data=lr.fit_transform(df)
print('Linear Regressor')
print(new_Data)

dt=IterativeImputer(estimator=DecisionTreeRegressor())
new_Data=dt.fit_transform(df)
print('Decision Tree Regressor')
print(new_Data)

rf=IterativeImputer(estimator=RandomForestRegressor(n_estimators=100))
new_Data=rf.fit_transform(df)
print("Random Forest Regressor")
print(new_Data)
```
