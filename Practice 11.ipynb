```python
data={'family_name':['Ali','Taimoor','Ahsan'],
     'Family_size':[4,8,3],
     'Income':[2000,1000,5000]}
df=pd.DataFrame(data)
df

def family_cat(x):
    if x<=3:
        return 'Small Family'
    elif x<=4:
        return 'Medium Family'
    else:
        return 'Big Family'

def income(x):
    if x<=1000:
        return 'Low income'
    elif x<=2000:
        return 'Medium income'
    else:
        return 'High Income'

df['Family_size_C']=df['Family_size'].apply(family_cat)
df['Income_C']=df['Income'].apply(income)
df
```

```python
import pandas as pd

# Sample data
data = {'family_name': ['Ali', 'Taimoor', 'Ahsan'],
        'Family_size': [4, 8, 3],
        'Income': [2000, 1000, 5000]}
df = pd.DataFrame(data)

# Define bins and labels for Family_size
family_bins = [0, 3, 4, float('inf')]  # 0-3, 4-5, 5 and above
family_labels = ['Small Family', 'Medium Family', 'Big Family']
df['Family_size_C'] = pd.cut(df['Family_size'], bins=family_bins, labels=family_labels)

# Define bins and labels for Income
income_bins = [0, 1000, 2000, float('inf')]  # 0-1000, 1001-2000, 2001 and above
income_labels = ['Low income', 'Medium income', 'High Income']
df['Income_C'] = pd.cut(df['Income'], bins=income_bins, labels=income_labels)

print(df)
```
